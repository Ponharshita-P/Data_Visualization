## MATPLOTLIB

1)It is a open source drawing library which supports rich drawing types.

2)It is used to draw 2D and 3D graphics.

3)We can understand our data easily by visualizing it with the help of matplotlib.

4)We can generate plots, histograms and many other charts with just a few lines of code.

Types of plots :
 
  * Bar Chart
  * Scatter Plot
  * Pie Chart
  * Histogram
  * Line Chart
  * Area Graph

```python
from matplotlib import pylab
```

```python
# Use Numpy to generate random data

import numpy as np

x = np.linspace(0, 10, 25)
y = x * x + 2
print(x)
print(y)
```

```python
print(np.array([x,y]).reshape(25,2))
```

```python
print(np.array([x,y]).reshape(2,25))
```

```python
pylab.plot(x,y,'r') #'r' stands for red
```

```python
# Drawing a subgraph

pylab.subplot(1,2,1) #(rows, columns, indexes)
pylab.plot(x,y,'r--')

pylab.subplot(1,2,2)
pylab.plot(x,y,'g*-')
```

### Operator Description

plt.figure() = Initializes subplot

fig.add_axes() = Initializes axes of the subplot 

a = fig.add_subplot(222) = Adds one subplot

fig,b = plt.subplots(nrows=3, ncols=2) = Adds more than 1 subplot

ax = plt.subplots(2,2) = Creates subplot

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()
axes = fig.add_axes([0.5, 0.1, 0.8, 0.8]) # Control the left, right, width, height of the canvas (from 0 to 1)
axes.plot(x,y,'r')
```

```python
# Draw subgraphs 
fig, axes = plt.subplots(nrows=1, ncols=2) #Submap is of 1 row and 2 columns
for ax in axes:
    ax.plot(x,y,'b')
```

```python
# Draw a picture or graph inside another  graph
fig = plt.figure()

#Control the left, right, width, height of the canvas (from 0 to 1)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #big axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #small axes

axes1.plot(x,y,'b')
axes2.plot(x,y,'r')
```

```python
fig = plt.figure(figsize=(16,9), dpi=300) #New graphic object
fig.add_subplot()
plt.plot(x,y,'r')
```
